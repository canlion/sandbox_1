```python
train_path = '/hdd/dataset/NYU_Depth_V2/data/nyu2_train/'
test_path = '/hdd/dataset/NYU_Depth_V2/data/nyu2__test/'
```

```python
list_ds = tf.data.Dataset.list_files(os.path.join(train_path, '*/*.jpg'))
```

```python
for f_path in list_ds.take(5):
    print(f_path.numpy())
```

```python
def get_data(f_path):
    x = tf.io.read_file(f_path)
    x = tf.image.decode_jpeg(x)
    y_f_path = tf.strings.regex_replace(f_path, 'jpg', 'png')
    y = tf.io.read_file(y_f_path)
    y = tf.image.decode_png(y)
    
    return x, y
```

```python
ds = list_ds.map(get_data)
```

```python
max_x_value = 0
max_y_value = 0
for i, (x, y) in enumerate(ds):
    if i % 100 == 0:
        print(i)
#     print(x.dtype, y.dtype)
#     plt.figure(figsize=(10, 10))
#     plt.subplot(121)
#     plt.imshow(y[..., 0])
#     y_norm = (y / tf.reduce_max(y)) * 10
#     plt.subplot(122)
#     plt.imshow(y_norm[..., 0])
#     plt.show()
#     print(tf.reduce_max(y_norm))
    max_x_value = max(max_x_value, tf.reduce_max(x))
    max_y_value = max(max_y_value, tf.reduce_max(y))
print(max_x_value, max_y_value)
```
